### Feature Encoding: Diagnostic Analysis

```python
diagnostic_ds
```

**Cardinality**:
- Low: unique values less than 10
- Medium: unique values between 10 and 20
- High: unique values greater than 20

```python
# Function: define the cardinality category
def determine_cardinality_of_func(x: int):
    if x < 10:
        return "Low"
    elif 10<=x<=20:
        return "Medium"
    else:
        return "High"
```

```python
# Identify features that are ordinal features 
diagnostic_cat_ds = diagnostic_ds[diagnostic_ds["DataTypes"] == object]
diagnostic_cat_ds.loc[:, ["CategoricalType"]] = diagnostic_cat_ds["NumDistinctValues"].apply(lambda x: "Binary" if x == 2 else "Nominal")
diagnostic_cat_ds.index = [i for i in range(diagnostic_cat_ds.shape[0])]

# Define the cardinality of each feature 
diagnostic_cat_ds.loc[:, ["Cardinality"]] = diagnostic_cat_ds["NumDistinctValues"].transform(determine_cardinality_of_func)
diagnostic_cat_ds
```

- Binary Features: Touchscreen, Msoffice
- Ordinal Features: None
- Nominal Features: all the rest

```python
# Remove the processor generation from the diagnostic dataset
diagnostic_cat_ds = diagnostic_cat_ds.drop(index=[3, 4], axis=0)
diagnostic_cat_ds.index = [idx for idx in range(diagnostic_cat_ds.shape[0])]
diagnostic_cat_ds

# Save the diagnostic data as version 2
diag_data_loader.save(("diagnostic_cat_ds", diagnostic_cat_ds), folder="diagnostics", version_nr=1)
```

For nominal features, we use One-Hot Encoding. It is useful to efficiently train as many models there are, such as regression models, knns, svms and tree-based models, like random forest, xgboost models and ligtgbm. The disadventage of OHE technique is it increases the dataset's dimensionality. 

**Solution for reducing dimensionality to preserve all relevant data:**
To reduce the dimensionality of one single dataset, we utilize the normalization technique that take one feature from the dataset and apply onehot encoding. These will provide multiple features, that will be considered to form one dataset. By encoding each feature, the dataset will be stored in a seperate folder called feature_datasets

### Feature Encoding: Nominal Features

```python
# OneHotEncoding
# Initialise the encoding object
# oht_encoder = FeatureEncoder(dataset1, "OneHotEncoding")
# feature_ds1 = oht_encoder.transform("Brand")
# feature_ds1

# Using OneHotEncoding: encode each feature one by one
nominal_feature_names = diagnostic_cat_ds[diagnostic_cat_ds["CategoricalType"] == "Nominal"].loc[:, ["FeatureNames"]].values.transpose()
nominal_feature_names = nominal_feature_names.flatten()

oht_encoder = FeatureEncoder(dataset1, "OneHotEncoding")
folder_loc = oht_encoder.feature_ds_folder_path

for idx, nominal_f_name in enumerate(nominal_feature_names):
    # Encode each feature using OneHotEncoding
    feature_data = oht_encoder.transform(nominal_f_name)

    # Save the encoded feature in the correct folder 
    oht_encoder.save_ds(nominal_f_name, file_idx=idx)
    print(f"Nominal Feature ({nominal_f_name}) is stored successfully in folder loc {folder_loc}")

# Using LabelEncoding:
laben_encoder = FeatureEncoder(dataset1, "OneHotEncoding")
folder_loc = laben_encoder.feature_ds_folder_path

for idx, nominal_f_name in enumerate(nominal_feature_names):
    # Encode each feature using OneHotEncoding
    feature_data = laben_encoder.transform(nominal_f_name)

    # Save the encoded feature in the correct folder 
    laben_encoder.save_ds(nominal_f_name, file_idx=idx)
    print(f"Nominal Feature ({nominal_f_name}) is stored successfully in folder loc {folder_loc}")
```
